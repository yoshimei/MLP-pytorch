```python
import numpy as np
import pickle, gzip
#import mlp
import matplotlib.pyplot as plt
import torch
import time
import cv2
```

```python
def openfile(path):
    f = open(path,"r", encoding='utf-8')
    lines = f.readlines()

    img_path=[]  
    classes=[]
    for line in lines:
        sl=line.strip().split(" ")
        img_path.append(sl[0]) 
        classes.extend(sl[-1].split(','))
    number = 1    
    zenbu = len(img_path)
    imgs = []

    print("loading images...")
    for j in img_path:
        
        #print(j)
        img  = cv2.imread(j)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img=cv2.resize(img,(50,50),interpolation=cv2.INTER_CUBIC)


        imgs.append(img)
        print(number,"/",zenbu,"  ", end="\n")
        number += 1
    print("\nDone!")
    return(imgs,img_path,classes)
```

```python
img_path=[]
classes=[]
imgs=[]
imgs,image_paths,classes=openfile("train.txt")
```

```python
imgs = np.array(imgs ).astype('float32')
imgs = np.array(imgs ).reshape(len(imgs),2500)
imgs /= 255
```

```python
aa = classes
bb = classes
print(imgs.shape)
bb = list(map(int, bb))
#classes = np.array(classes ).reshape(len(classes))
print(bb[0])
```

```python
from keras.utils import np_utils

bb = np_utils.to_categorical(bb)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add( Dense( input_dim = 50*50          #  28*28=784 
                , units = 256                   #  256  (, , )
                , kernel_initializer = 'normal' #  normal  weight  bias 
                , activation = 'relu'           #  relu 
                )
         )
```

```python
model.add( Dense( units = 50                    #  10  ( 0 ~ 9)
                , kernel_initializer = 'normal' #  normal  weight  bias 
                , activation = 'softmax'        #  softmax  (softmax , )
                )
         )
```

```python
model.compile( loss='categorical_crossentropy' #  Loss   categorical_crossentropy
             , optimizer = 'adam'              #  Optimizer   adam
             , metrics = ['accuracy']          #  Model   accuracy
             )
```

```python
history = model.fit(               # ,  history
          x = imgs #   Features  (mnist  60000 )
        , y = bb   #   Label     (mnist  60000 )
        , validation_split = 0.2   #           (60000*0.2=12000 , 60000*0.8=48000 )
        , epochs = 50              #               ( 10 ,  , , )
        , batch_size = 1000        #   ( 100 , , , )
        , verbose = 2              #           (0: , 1: , 2: )
)
```

```python
import matplotlib.pyplot as plot  # plot 

def train_history_graphic( history       # 
                         , history_key1  #  1 ( loss, acc, val_loss, val_acc )
                         , history_key2  #  2 ( loss, acc, val_loss, val_acc )
                         , y_label       # Y 
                         ) :
    #  1
    plot.plot( history.history[history_key1] )

    #  2
    plot.plot( history.history[history_key2] )

    # 
    plot.title( 'train history' )

    # X 
    plot.xlabel( 'epochs' )

    # Y 
    plot.ylabel( y_label )

    # 
    # ( 1 , , )
    # ( 2 , upper , lower , left , right )
    plot.legend( ['train', 'validate']
               , loc = 'upper left'
               )

    # 
    plot.show()

####################
#  train history 
####################
train_history_graphic( history, 'accuracy', 'val_accuracy', 'loss' )
```

```python
img_path_test=[]
classes_test=[]
imgs_test=[]
imgs_test,image_paths,classes_test=openfile("test.txt")
imgs_test = np.array(imgs_test ).astype('float32')
imgs_test = np.array(imgs_test ).reshape(len(imgs_test),2500)
imgs_test /= 255
```

```python

print(imgs_test.shape)
classes_test = list(map(int, classes_test))
#classes = np.array(classes ).reshape(len(classes))
print(classes_test[0])
```

```python

####################
# MLP  Data 
# ,  Test Label  One-Hot Encoding
####################
classes_test = np_utils.to_categorical(classes_test)

####################
#  Model 
####################
eval = model.evaluate( imgs_test, classes_test )

#  Model 
print( 'loss =', eval[0] )
print( 'accuracy =', eval[1] )
```
