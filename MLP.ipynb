```python
import numpy as np
import pickle, gzip
#import mlp
import matplotlib.pyplot as plt
import torch
import time
import cv2
```

```python
def openfile(path):
    f = open(path,"r", encoding='utf-8')
    lines = f.readlines()

    img_path=[]  
    classes=[]
    for line in lines:
        sl=line.strip().split(" ")
        img_path.append(sl[0]) 
        classes.extend(sl[-1].split(','))
    number = 1    
    zenbu = len(img_path)
    imgs = []

    print("loading images...")
    for j in img_path:
        
        #print(j)
        img  = cv2.imread(j)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img=cv2.resize(img,(50,50),interpolation=cv2.INTER_CUBIC)


        imgs.append(img)
        print(number,"/",zenbu,"  ", end="\n")
        number += 1
    print("\nDone!")
    return(imgs,img_path,classes)
```

```python
img_path=[]
classes=[]
imgs=[]
imgs,image_paths,classes=openfile("train.txt")
```

```python
imgs = np.array(imgs ).astype('float32')
imgs = np.array(imgs ).reshape(len(imgs),2500)
imgs /= 255
```

```python
aa = classes
bb = classes
print(imgs.shape)
bb = list(map(int, bb))
#classes = np.array(classes ).reshape(len(classes))
print(bb[0])
```

```python
from keras.utils import np_utils

bb = np_utils.to_categorical(bb)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add( Dense( input_dim = 50*50          # 輸入層有 28*28=784 個神經元
                , units = 256                   # 隱藏層有 256 個神經元 (值越大, 訓練越精準, 相對訓練時間也越久)
                , kernel_initializer = 'normal' # 使用 normal 初始化 weight 權重與 bias 偏差值
                , activation = 'relu'           # 使用 relu 激活函數
                )
         )
```

```python
model.add( Dense( units = 50                    # 輸出層有 10 個神經元 (因為數字只有 0 ~ 9)
                , kernel_initializer = 'normal' # 使用 normal 初始化 weight 權重與 bias 偏差值
                , activation = 'softmax'        # 使用 softmax 激活函數 (softmax 值越高, 代表機率越大)
                )
         )
```

```python
model.compile( loss='categorical_crossentropy' # 設定 Loss 損失函數 為 categorical_crossentropy
             , optimizer = 'adam'              # 設定 Optimizer 最佳化方法 為 adam
             , metrics = ['accuracy']          # 設定 Model 評估準確率方法 為 accuracy
             )
```

```python
history = model.fit(               # 訓練的歷史記錄, 會會回傳到指定變數 history
          x = imgs # 設定 圖片 Features 特徵值 (mnist 提供 60000 筆資料)
        , y = bb   # 設定 圖片 Label    真實值 (mnist 提供 60000 筆資料)
        , validation_split = 0.2   # 設定 有多少筆驗證         (60000*0.2=12000 筆驗證, 60000*0.8=48000 筆訓練)
        , epochs = 50              # 設定 訓練次數             (值 10 以上,  值越大, 訓練時間越久, 但訓練越精準)
        , batch_size = 1000        # 設定 訓練時每批次有多少筆 (值 100 以上, 值越大, 訓練速度越快, 但需記憶體要夠大)
        , verbose = 2              # 是否 顯示訓練過程         (0: 不顯示, 1: 詳細顯示, 2: 簡易顯示)
)
```

```python
import matplotlib.pyplot as plot  # plot 可以視為畫布

def train_history_graphic( history       # 資料集合
                         , history_key1  # 資料集合裡面的來源 1 (有 loss, acc, val_loss, val_acc 四種)
                         , history_key2  # 資料集合裡面的來源 2 (有 loss, acc, val_loss, val_acc 四種)
                         , y_label       # Y 軸標籤文字
                         ) :
    # 資料來源 1
    plot.plot( history.history[history_key1] )

    # 資料來源 2
    plot.plot( history.history[history_key2] )

    # 標題
    plot.title( 'train history' )

    # X 軸標籤文字
    plot.xlabel( 'epochs' )

    # Y 軸標籤文字
    plot.ylabel( y_label )

    # 設定圖例
    # (參數 1 為圖例說明, 有幾個資料來源, 就對應幾個圖例說明)
    # (參數 2 為圖例位置, upper 為上面, lower 為下面, left 為左邊, right 為右邊)
    plot.legend( ['train', 'validate']
               , loc = 'upper left'
               )

    # 顯示畫布
    plot.show()

####################
# 顯示 train history 準確率圖表
####################
train_history_graphic( history, 'accuracy', 'val_accuracy', 'loss' )
```

```python
img_path_test=[]
classes_test=[]
imgs_test=[]
imgs_test,image_paths,classes_test=openfile("test.txt")
imgs_test = np.array(imgs_test ).astype('float32')
imgs_test = np.array(imgs_test ).reshape(len(imgs_test),2500)
imgs_test /= 255
```

```python

print(imgs_test.shape)
classes_test = list(map(int, classes_test))
#classes = np.array(classes ).reshape(len(classes))
print(classes_test[0])
```

```python

####################
# MLP 輸出層的每筆 Data 都是一維陣列
# 所以, 需將 Test Label 資料轉換為 One-Hot Encoding
####################
classes_test = np_utils.to_categorical(classes_test)

####################
# 使用測試資料評估 Model 的【損失函數誤差值】與【準確率】
####################
eval = model.evaluate( imgs_test, classes_test )

# 顯示 Model 評估測試資料的【損失函數誤差值】與【準確率】
print( 'loss =', eval[0] )
print( 'accuracy =', eval[1] )
```
